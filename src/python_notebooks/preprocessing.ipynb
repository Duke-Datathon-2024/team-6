```python
df_demographics = pd.read_csv('../../data/dataset_duke_datathon_team6.csv')
df_demographics
```

```python
# Merge the datasets
df = pd.merge(df_cohort, df_demographics, on='stay_id', how='left')
df
```

```python
df.shape
```

```python
df.columns
```

```python
# Mapping to group locations into categories
location_mapping = {
    'HOME': 'Home or equivalent',
    'HOME HEALTH CARE': 'Home or equivalent',
    'ASSISTED LIVING': 'Home or equivalent',
    'AGAINST ADVICE': 'Home or equivalent',
    'DIED': 'Death or hospice',
    'HOSPICE': 'Death or hospice',
    'ACUTE HOSPITAL': 'Other hospital or equivalent',
    'CHRONIC/LONG TERM ACUTE CARE': 'Other hospital or equivalent',
    'OTHER FACILITY': 'Other hospital or equivalent',
    'PSYCH FACILITY': 'Other hospital or equivalent',
    'HEALTHCARE FACILITY': 'Other hospital or equivalent',
    'REHAB': 'Rehab',
    'SKILLED NURSING FACILITY': 'Skilled nursing facility'
}
```

```python
df['disposition'] = df.discharge_location.map(location_mapping)
df['disposition']
```

```python
df['disposition'].value_counts()
```

```python
df['race_group'].value_counts()
```

```python
df['race'].value_counts()
```

```python
# Mapping race group into categories White and Non-white
white_race_dict = {
    'White': 'White',
    'Black': 'Non-white',
    'Hispanic': 'Non-white',
    'Asian': 'Non-white',
    'Other': 'Non-white'
}
```

```python
df['white_race'] = df['race_group'].map(white_race_dict)
```

```python
df['race_group'].value_counts()
```

```python
df['white_race'].value_counts()
```

```python
df['los_icu_survivors'] = df['los_icu'].where(df['mortality_90'] == 0)
```

```python
df['los_icu_expired'] = df['los_icu'].where(df['mortality_90'] == 1)
```

```python
df.shape
```

```python
columns_to_keep = [
      'subject_id', 'hadm_id', 'stay_id', 'gender', 
      'dod', 'admittime','dischtime', 
      'los_hospital', 'admission_age', 'white_race', 'hospital_expire_flag', 
       'hospstay_seq', 'first_hosp_stay', 'icu_intime', 'icu_outtime', 
       'los_icu', 'icustay_seq', 'first_icu_stay', 'race_group',
       'sex_female','los_icu_1', 'language',
       'eng_prof', 'insurance', 'private_insurance',
       'mortality_in', 'mortality_90', 'disposition',
       'white_race', 'los_icu_survivors', 'los_icu_expired',
       'anchor_age', 'charlson_comorbidity_index',
       'discharge_location', 'SOFA'
       ]
```

```python
df = df[columns_to_keep]
df
```
